```python
def extract_file_data(root_path, folder_name) -> dict:
    documents = {}

    documents_folder = os.path.join(root_path, folder_name)

    if not os.path.exists(documents_folder):
        return documents
    
    files = os.listdir(documents_folder)
    for file in files:
        if file.endswith(".txt"):
            with open(os.path.join(documents_folder, file), 'r', encoding='utf-8') as f:
                data = f.read()
            documents[file] = data
    return documents
```

```python
root_path = r"C:\D\NLP_Project\Own_creation\Customer_Support_chat_bot"
folder_name = "Documents"
documents = extract_file_data(root_path, folder_name)
# print(documents)
```

```python
api_key = os.getenv("COHERE_API_KEY") 
co = cohere.Client(api_key=api_key)
```

```python
all_docs = list(extract_file_data(root_path, folder_name).values()) 
doc_embeddings = co.embed(texts=all_docs, model='embed-english-v3.0', input_type='search_document')
doc_embeddings
```

```python
query = ["How do I troubleshoot slow internet"]
query_embeddings = co.embed(texts=query, model='embed-english-v3.0', input_type='search_query', embedding_types=['ubinary'])
query_embeddings
```
